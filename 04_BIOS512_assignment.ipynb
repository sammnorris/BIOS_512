```r
library('tidyverse')
```

```r
turbine_df <- read_csv('https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2020/2020-10-27/wind-turbine.csv')
```

```r
turbine_df |> head()
```

Each row in the table represents a single turbine.

Make a bar chart üìä **sorted by frequency** of turbine count per province territory. (Hint: `fct_infreq` will sort your factor levels by frequency for subsequent plotting)

```r
turbine_sorted = turbine_df |> mutate(province_sorted=province_territory |> fct_infreq())

options(repr.plot.height=8, repr.plot.width=12)

plot1=ggplot(turbine_sorted, aes(x=province_sorted)) + geom_bar() +
    theme(axis.text.x=element_text(angle=45, vjust=0.5)) +
    labs(x="Province", y="Turbines")

plot1
```

Make a bar chart **sorted by frequency** of turbine count per manufacturer. **Group the low frequency manufacturers into an `Other` category**.

**Bonus:** Can you force the `Other` category to be the last bar? 

```r
table(turbine_df$manufacturer)
```

```r
turbine_sorted2 = turbine_df |> mutate(manufacturer_lumped = manufacturer |> fct_lump_min(40) |>fct_infreq() |> fct_relevel("Other", after=Inf))

table(turbine_sorted2$manufacturer_lumped)
```

```r
plot2=ggplot(turbine_sorted2, aes(x=manufacturer_lumped)) + geom_bar() +
    theme(axis.text.x=element_text(angle=45, vjust=0.5)) +
    labs(x="Manufacturer", y="Turbines")

plot2
```
